```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

#Data Viz lib
import plotly.express as px
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt


# Set the theme for charts
import plotly.io as pio
pio.templates.default = "simple_white"

#Data URL
data_url='/kaggle/input/known-brands-slogans-and-risk-assessment/Slogans.csv'

#Read the data into df frame
df=pd.read_csv(data_url)
print('------Shape of the dataFrame----------')
#No of row and columns
print('No. of Row: {}, Columns: {}'.format(df.shape[0],df.shape[1]))
print('\n-------Columns Name---------')
print(df.columns)
```

## Comman words in the Brand Slogans
*Typically include : Better, good, now, life, make*

```python
comment_words = '' 
stopwords = set(STOPWORDS) 

lst = df['Slogan'][:].values.tolist()
for val in lst: 
    val = str(val) 
    tokens = val.split() 

    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
      
    comment_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 600, height = 400, background_color ='black', stopwords = stopwords, min_font_size = 10).generate(comment_words) 
                       
plt.figure(figsize = (20, 6), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()
```

```python
df.sort_values(by='simillar_trademarks',ascending=False,inplace=True)
dfo=df[:10]
fig = px.bar(dfo, x="Brand", 
             y="simillar_trademarks",
             title='Top ten brans who has Similar Trademarks',
            hover_data=['identical_trademarks', 'simillar_trademarks'],)
fig.show()
```

```python
df.sort_values(by='identical_trademarks',ascending=False,inplace=True)
dfo=df[:10]
fig = px.bar(dfo, x="Brand", 
             y="identical_trademarks",
             title='Top ten brands who has Identical Trademarks',
            hover_data=['identical_trademarks', 'simillar_trademarks'],)
fig.show()
```
